```python
import gzip
from Bio import SeqIO
```

```python
with gzip.open(filename='../data/uniprot_sprot.fasta.gz', mode='rt') as handle:
    records = list(SeqIO.parse(handle, 'fasta'))
    print(f"Total counts: {len(records)}")
    
    for i, r in enumerate(records[:3]):
        print(f"\n---Record {i+1}---")
        print(f"ID: {r.id}")
        print(f"Description: {r.description}")
        print(f"Length: {len(r.seq)}")
        print(f"Sequence (first 60 aa): {r.seq[:60]}")
```
